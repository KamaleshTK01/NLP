```python
# Day 2
# Agenda:
#   Text processing -> Words-> Vectors
#     a. OHE (One Hot Encoding)
#     b. Bag of words (BOW) and ngrams
#     c. TF-IDF (Term Frequency - Inverse Document Frequency)
#     d. Word2vec

# Basic terminologies used in NLP:
#   1. Corpus  => Paragraph
#   2. Documents  => Sentence
#   3. Vocabulary => 10k Unique words
#   4. Words  => Word

# Sentiment Analysis

# Text   --> output
# D1 The food is good   1
# D2 The food is bad    0
# D3 Pizza is amazing   1
# D4 Burger is bad      0

# Dataset --> Text preprocessing - 1 --> Text preprocessing - 2 --> Converting the words into vectors
#             Tokenization, Lowering     Stemming, Lemitization,        BOW, TF-IDF, Word2vec
#             the case of the words         Stop words

# I. One Hot Encoding:
#         [A man eat food
#         Cat eat food
#         People watch krish youtube channel]

#         Vocabulary --> 10 [A man eat food cat people watch krish youtube channel]
#         D1 is Document-1

#         Line one in the paragraph is D1:
#           [[1000000000]
#           [0100000000]
#           [0010000000]
#           [0001000000]]  => OHE Format

#         Adavantages:
#           Simple to implement.
#           Initiative

#         DisAdvantages:
#           Create a Sparse Matrix.
#           Out of Vocabulary (OOV)
#           Not fixed size of sentence
#           Semantic meaning between the word is not captured.
# II. Bag of Words:
#         D1 --> He is a good boy
#         D2 --> She is a good girl
#         D3 --> Boy and girl are good

#         After applying the stop words:
#           D1 --> good boy
#           D2 --> Good girl
#           D3 --> Boy girls\ good
#         1. Finding out the vocabulary:
#             Pronouns are not considered by applying stop words.
#                       good, boy, girl
#             Frequency - 3    2    2

#                  f1    f2   f3
#             --> Good  boy  girl  O/P
#           Doc-1  1     1    0
#           Doc-2  1     0    1
#           Doc-3  1     1    1

#           Advantages:
#             Simple and intutive

#           Disadvantages:
#             Sparsity issue is not getting fixed.
#             OOV
#             Ordering of the words has completely changed.
#             Semantic meaning is lost.

# Cosine Similarity = 1 - similarity
# Similarity cos angle between the data points

# Ngrams:
#         f1    f2    f3      f4        f5
#       Good  Boy   Girls  Goodboy  Goodgirl
# Snt-1  1      1    0       1         0
# Snt-2  1      0    1       0         1
# Snt-3  1      1    1       0         0

# Krish Eats Food
#   How many bigrams ?
#     1. Krish Eats
#     2. Eats food

# I am not feeling well
#   How many diffrent trigrams?
#     1. I am not
#     2. am not feeling
#     3. not feeling well

# Krish is not feeling well
#   (1,3) = Uni to trigram
#   f1     f2    f3    f4      f5    f6   .......... N grams
#   Krish  is   not  feeling  well  krish is   ....... Ngrams




```

```python
import re
corpus = []
for i in range(len(sentences)):
  review = re.sub('[^a-zA-Z]',' ',sentences[i])
  review = review.lower()
  corpus.append(review)
```

```python
corpus
```

```python
##stemming
nltk.download('stopwords')
for i in corpus:
  words = nltk.word_tokenize(i)
  for word in words:
    if word not in set(stopwords.words('english')):
      print(stemmer.stem(word))
```

```python
##lemmatize
nltk.download('stopwords')
for i in corpus:
  words = nltk.word_tokenize(i)
  for word in words:
    if word not in set(stopwords.words('english')):
      print(lemmatizer.lemmatize(word))
```
